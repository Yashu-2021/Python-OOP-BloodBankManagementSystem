# OOP-Quiz-1

The following assumptions have been made in this blood donation program:
•There is no universal recipient blood group
•There is no universal donor blood group
•Blood group AB- can donate to B- as well
•Blood group AB+ can donate to B+ as well

```python
#Blood Donation Management system

class user:
    name=''
    group=''
    area=''
    cont=''

class admin:
    name=''
    adminid=''

def detail_user():
    user.name=input("please Enter Name: ")
    user.group=input("Please Enter Blood group: ")
    user.area=input("Please Enter Age : ")
    user.cont=input("Please Enter Contact number: ")
    status=input("Donor[d] or Recipient[r]: ")
    return status

def request():
    file=open('donors.txt','r')
    flag=0
    for line in file:
        if user.group in line:
            if user.area in line:
                flag=1
                print("The details of donor: ")
                print(line)
    file.close()
    if flag==0:
        print("No donor available")

def don():
    file=open('donors.txt','a')
    L=[user.name,'              ',user.group,
            '                ',user.area,'                  ',user.cont]
    file.writelines(L)
    file.close()

def rec():
    lvl=input("Is this an emergency?[yes/no] ")
    if lvl=='yes':
        request()
    elif lvl=='no':
        file=open('donors.txt','a')
        file.write("There is a request :")
        L=[user.name,'              ',user.group,
            '                ',user.area,'                  ',user.cont]
        file.writelines(L)
        file.close()
        print("Your request for blood has been entered")
    else: 
        print("Incorrect choice")

def use():
    b=detail_user()
    if b=='d':
        don()
    elif b=='r':
        rec()
    else:
        print("Invalid input")

#valid admin IDs are 8564, 7392, 3682, 4679

def detail_admin():
    admin.name=input("Enter your name: ")
    admin.adminid=input("Enter the Admin ID: ")
    return

def check():
    if admin.adminid=='8564'or '7392' or '3682' or '4679':
        func()
    else:
        print("Invalid admin ID")

def dets():
    user.name=input("please Enter Name: ")
    user.group=input("Please Enter Blood group: ")
    user.area=input("Please Enter Age : ")
    user.cont=input("Please Enter Contact number: ")


def func():
    f=input("View details or Add details or Send request? [view/add/request]")
    if f=='view':
        file=open('donors.txt','request')
        for line in file:
            print(line)
        file.close()
    elif f=='add':
        dets()
        don()
    elif f=='request':
        dets()
        request()

def adm():
    detail_admin()
    check()

p=input("Are you a user or an administrator?[user/adm]")
if p == 'user':
    use()
elif p == 'adm':
    adm()
else:
    print("Wrong choice") 
```
